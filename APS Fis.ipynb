```python
import math
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.integrate import odeint
from IPython.display import Image
```

```python
Image("diag.jpeg")
```

```python
t = np.arange(0, 3, 0.01)

l = 1.24 #raio de 1,24 metros
g = 9.8

def func(solucao, t):
    T= solucao[0]
    o = solucao[1]
    dtdt = o  
    dodt = (-g*math.sin(T))/l
  
    return dtdt,dodt
```

```python

ai = [math.radians(30),0]  # angulo inicial de aproximadamente 30 graus

solucao = odeint(func,ai,t)

x = []
y = []

for teta in solucao[:,0]:
    x.append(math.cos(teta)*l)
    y.append(math.sin(teta)*l)
    
plt.plot(y, x)
plt.title('TrajetÃ³ria PÃªndulo com 30Â° de abertura')
plt.xlabel('x')
plt.ylabel('y')
plt.axis([min(y),max(y),max(x),min(x)])
plt.axis('equal')
plt.grid()
plt.show()

###GRAFICOS FORA DE ESCALA
```

```python
ai = [math.radians(10),0] #angulo aproxiamadanete de 10graus

solucao = odeint(func,ai,t)

x = []
y = []

for teta in solucao[:,0]:
    x.append(math.cos(teta)*l)
    y.append(math.sin(teta)*l)
    
plt.plot(y, x)
plt.title('TrajetÃ³ria PÃªndulo com 10Â° de abertura')
plt.xlabel('x')
plt.ylabel('y')
plt.axis([min(y),max(y),max(x),min(x)])
plt.axis('equal')
plt.grid()
plt.show()

###GRAFICOS FORA DE ESCALA
```
